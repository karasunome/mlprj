```python
import matplotlib.pyplot as plt

fp = open("sim_results.txt")

epochs = []
train_accs = []
train_losses = []
val_accs = []
val_losses = []
lrs = []

for line in fp.readlines():
    epochs.append(int(line[:2]))
    lrs.append(float(line[3:9]))
    train_losses.append(float(line[10:17]))
    train_accs.append(float(line[17:24]))
    val_losses.append(float(line[25:32]))
    val_accs.append(float(line[33:40]))
```

```python
plt.plot(val_accs, '-x')
plt.plot(train_accs, '-x')
plt.xlabel('epoch')
plt.ylabel('accuracies')
plt.legend('vt')
plt.title('Accuracy vs. No. of epochs')
```

```python
plt.plot(val_losses, '-x')
plt.plot(train_losses, '-x')
plt.xlabel('epoch')
plt.ylabel('accuracies')
plt.legend('vt')
plt.title('Accuracy vs. No. of epochs')
```

```python
plt.plot(lrs, '-x')
plt.xlabel('epoch')
plt.ylabel('learning rate')
plt.title('Learning rate and No of epochs')
```
